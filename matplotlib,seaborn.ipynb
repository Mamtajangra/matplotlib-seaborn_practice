# bar chart

```python
x = np.array([10,13,16,20,34,56,67])
y = np.array(["A","B","C","D","E","F","G"])
plt.bar(x,y,color = "red",width = 10)
plt.title("country activities")
plt.ylabel("population")
plt.xlabel("activities")
plt.grid()
plt.show()
```

# horizontal bar chart

```python
x = np.array([10,13,16,20,34,56,67])
y = np.array(["A","B","C","D","E","F","G"])
plt.barh(x,y,color = "hotpink",height= 10)
plt.title("country activities")
plt.ylabel("population")
plt.xlabel("activities")
plt.grid()

plt.show()
```
